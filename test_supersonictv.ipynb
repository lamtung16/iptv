```python
import requests
import csv
from requests.exceptions import RequestException

output_file = "scan_supersonic_results.csv"
user = "006779"
password = "439866"
base_url = f"http://supersonictv.live:8080/{user}/{password}/"

# Open the file and write header
with open(output_file, mode='w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(["ID", "Status", "Full_URL"])
    print(f"Starting scan... logging ONLY working channels to {output_file}")
    for test_id in range(10000, 100000):
        url = f"{base_url}{test_id}"
        try:
            with requests.get(url, stream=True, timeout=1.5) as r:
                if r.status_code == 200:
                    print(f"[WORKING] ID: {test_id}")
                    writer.writerow([test_id, r.status_code, url])
                    f.flush()
                else:
                    print(f"Scanning: {test_id} (Status: {r.status_code})", end="\r")
                    
        except RequestException:
            continue

print("\nScan complete.")
```
