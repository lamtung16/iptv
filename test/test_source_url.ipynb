```python
import requests

def filter(BASE_URL, USERNAME, PASSWORD, field, value):
    params = {
        "username": USERNAME,
        "password": PASSWORD,
        "action": "get_live_streams"
    }
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
    }
    response = requests.get(BASE_URL + "/player_api.php", params=params, headers=headers, timeout=5)
    response.raise_for_status()
    streams = response.json()
    output_streams = [s for s in streams if value.lower() in str(s.get(field).lower())]
    return output_streams
```
