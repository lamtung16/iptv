```python
import pandas as pd
import requests
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

```python
def check_url(url):
    if pd.isna(url) or str(url).strip() == "":
        return "EMPTY"
    try:
        with requests.get(
            url,
            timeout=(5, 15),
            stream=True,
            allow_redirects=True,
            verify=False
        ) as response:
            if response.status_code == 200:
                chunk = next(response.iter_content(1024), None)
                if chunk:
                    return "WORKING"
            return "ERROR"

    except Exception as e:
        return f"ERROR"
```

```python
for channel in ['p', 'bu', 'it', 'l1', 'la', 'c', 'e']:
    for id in ['', 1, 2, 3, 4, 5]:
        url = f"https://lb.buzztip.site/{channel}{str(id)}/index.m3u8"
        print(f"{channel}{str(id):1s} - {url:40s} - {check_url(url)}")
```
