```python
import requests

def filter(BASE_URL, USERNAME, PASSWORD, field, value):
    params = {
        "username": USERNAME,
        "password": PASSWORD,
        "action": "get_live_streams"
    }
    headers = {
        "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36"
    }
    response = requests.get(BASE_URL + "/player_api.php", params=params, headers=headers, timeout=5)
    response.raise_for_status()
    streams = response.json()
    output_streams = [s for s in streams if value in str(s.get(field))]
    return output_streams
```
