```python
# inline {host} {username}:{password} {active_cons}/{max_connections} {mmm}{dd} {yyyy} {Status} {url} {support types} {location}
import re
import csv

input_file = "input.txt"
output_file = "output.csv"

with open(input_file, "r", encoding="utf-8") as f, open(output_file, "w", newline="", encoding="utf-8") as out:
    writer = csv.writer(out)

    # CSV header
    writer.writerow([
        "host", "username", "password", "status",
        "exp_date", "active_cons", "max_connections",
        "url", "location"
    ])

    for line in f:
        line = line.strip()
        if not line:
            continue

        # split by 2+ spaces
        parts = re.split(r"\s{2,}", line)

        try:
            host = parts[0]

            username, password = parts[1].split(":", 1)

            active_cons, max_connections = parts[2].split("/", 1)

            exp_date = f"{parts[3]} {parts[4]}"   # mmmdd yyyy

            status = parts[5]

            url = parts[6]

            location = parts[-1]

            row = [
                host, username, password, status,
                exp_date, active_cons, max_connections,
                url, location
            ]

            writer.writerow(row)

            # print each CSV line so you see it immediately
            print(",".join(row))

        except Exception as e:
            print("Skipped line:", line)
```

```python
import re
import csv

input_file = "input.txt"
output_file = "output.csv"

pattern = re.compile(
    r"├─𝙷𝚘𝚜𝚝 = (?P<host>.+?)\n"
    r"├─𝚁𝚎𝚊𝚕 = .+\n"
    r"├─𝙿𝚘𝚛𝚝 = .+\n"
    r"├─𝚄𝚜𝚎𝚛 = (?P<username>.+?)\n"
    r"├─𝙿𝚊𝚜𝚜 = (?P<password>.+?)\n"
    r"├─𝙴𝚡𝚙\.?\s*= .+\n"
    r"├─𝙰𝚌𝚝\s*𝙲𝚘𝚗 = (?P<active_cons>.+?)\n"
    r"├─𝙼𝚊𝚡\s*𝙲𝚘𝚗 = (?P<max_connections>.+?)",
    re.MULTILINE
)

with open(input_file, "r", encoding="utf-8") as f:
    data = f.read()

matches = pattern.finditer(data)

rows = []
for m in matches:
    rows.append([
        m.group("host"),
        m.group("username"),
        m.group("password"),
        m.group("active_cons"),
        m.group("max_connections")
    ])

with open(output_file, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["host", "username", "password", "active_cons", "max_connections"])
    writer.writerows(rows)

print(f"Exported {len(rows)} rows to {output_file}")
```

```python
import csv
import re

input_file = "input.txt"
output_file = "output.csv"

# Regex to capture host, username, password
pattern = re.compile(r"(https?://[^/]+)/get\.php\?username=([^&]+)&password=([^&]+)")

rows = []

with open(input_file, "r", encoding="utf-8", errors="ignore") as f:
    for line in f:
        line = line.strip()
        match = pattern.search(line)
        if match:
            host = match.group(1)
            username = match.group(2)
            password = match.group(3)
            rows.append([host, username, password])

# Write to CSV
with open(output_file, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["host", "username", "password"])
    writer.writerows(rows)

print(f"Saved {len(rows)} entries to {output_file}")
```
