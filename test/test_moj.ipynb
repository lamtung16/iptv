```python
import requests
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

```python
def check_url(url):
    try:
        with requests.get(
            url,
            headers=None,
            timeout=(2),
            stream=True,
            allow_redirects=True,
            verify=False
        ) as response:
            if response.status_code == 200:
                chunk = next(response.iter_content(1024), None)
                if chunk:
                    return "WORKING"
            return "ERROR"

    except Exception:
        return "ERROR"
```

```python
for channel in range(6128300, 6128700, 1):
    url = f"https://stream.decentdoubts.net/{channel}/index.m3u8"
    if check_url(url) == "WORKING":
        print(f"\nWORKING: {url}")
    print(f"Checking channel: {channel}", end="\r", flush=True)
```
