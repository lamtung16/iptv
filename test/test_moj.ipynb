```python
import pandas as pd
import requests
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

```python
def check_url(url):
    if pd.isna(url) or str(url).strip() == "":
        return "EMPTY"

    headers = {
        "Referer": "https://roxiestreams.info/",
        "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36"
    }

    try:
        with requests.get(
            url,
            headers=None,
            timeout=(5, 15),
            stream=True,
            allow_redirects=True,
            verify=False
        ) as response:
            if response.status_code == 200:
                chunk = next(response.iter_content(1024), None)
                if chunk:
                    return "WORKING"
            return "ERROR"

    except Exception:
        return "ERROR"
```

```python
for channel in range(100000):
    url = f"https://stream.decentdoubts.net/{channel}/index.m3u8"
    if check_url(url) == "WORKING":
        print(url)
```
