```python
import requests
```

```python
def find_match(keyword):
    url = 'https://streamed.pk/api/matches/football'
    matches = requests.get(url).json()
    target_match = next(m for m in matches if keyword.lower() in m.get('id', '').lower())
    source = target_match.get('sources', [])[0]
    stream_url = f"https://streamed.pk/api/stream/{source['source']}/{source['id']}"
    stream_res = requests.get(stream_url)
    return stream_res.json()[0]['embedUrl']
```

```python
find_match("manchester-united")
```
