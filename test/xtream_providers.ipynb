```python
import requests
from datetime import datetime

def get_xtream_account_info(host, username, password):
    # Construct the API URL
    api_url = f"{host}/player_api.php?username={username}&password={password}"
    print()
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/110.0.0.0 Safari/537.36'
        }
        response = requests.get(api_url, timeout=5, headers=headers)
        response.raise_for_status()  # Check for HTTP errors

        data = response.json()

        # Check if the credentials are valid
        if "user_info" in data:
            user_info = data["user_info"]
            server_info = data["server_info"]
            auth = user_info.get("auth", 0)

            if auth == 1:
                exp_timestamp = user_info.get("exp_date")

                # Handle cases where exp_date might be None or "null"
                if exp_timestamp and exp_timestamp != "null":
                    # Convert Unix timestamp to readable date
                    expiry_date = datetime.fromtimestamp(int(exp_timestamp))
                    readable_date = expiry_date.strftime('%Y-%m-%d %H:%M:%S')
                else:
                    readable_date = "Unlimited / No expiry"

                print(f"--- Account Info for {host}/{username} ---")
                print(f"Status: {user_info.get('status')}")
                print(f"Created Date: {datetime.fromtimestamp(int(user_info.get("created_at"))).strftime('%Y-%m-%d %H:%M:%S')}")
                print(f"Expiration Date: {readable_date}")
                print(f"Active Connections: {user_info.get('active_cons')}/{user_info.get('max_connections')}")
                print(f"Url: {server_info.get("url")}")
                print(f"Https port: {server_info.get("https_port")}")
                print(f"Protocol: {server_info.get("server_protocol")}")
                print(f"Rtmp Port: {server_info.get("rtmp_port")}")
            else:
                print("Error: Authentication failed. Check username/password.")
        else:
            print("Error: Unexpected response format.")

    except requests.exceptions.RequestException as e:
        print(f"Connection Error: {e}")
```
