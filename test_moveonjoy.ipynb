```python
import requests
from requests.exceptions import RequestException

for test_id in range(1, 101):
    url = f"https://fl{test_id}.moveonjoy.com/TENNIS_CHANNEL/index.m3u8"
    try:
        r = requests.get(url, stream=True, timeout=2)
        print(f"{test_id:4d}", r.status_code)
    except RequestException as e:
        print(f"{test_id:4d}", type(e).__name__)
```
